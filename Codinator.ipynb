```python
#Storing the alphabet and numbers for use with decoding
alph_ref = 'abcdefghijklmnopqrstuvwxyz0123456789'

#Used in the codinator function to get a key either from the user or generates a random one
def grab_key():
    
    #Has the user decide if they want to make a key or get a random one by inputing 1 or 2
    user_opt = input('To input your own key type 1, for a random key type 2: ')
    
    #Creates a blank string for use in a while loop
    final_key = ''
    
    #Generates a random 36 character key using a list of acceptable characters, then adds it to final_key
    if user_opt == '2':
        char_opt = 'abcdefghijklmnopqrstuvwxyz0123456789+=-_\|:;<>,."?/!@#$%^&*)(}{][`~'
        char_list = list(char_opt)
        for n in range(36):
            final_key += char_list.pop(rand.randrange(len(char_list)))
        print('Key: ' + final_key)
        return(final_key)
    
    #Loops while final_key is blank, lets the user input a key, continuing until it meets the critera
    #of having 36 unique characters. Once the critera are met, stores the input in final_key,
    #ending the loop
    elif user_opt == '1':
        while final_key == '':
            user_key = input('Please enter a 36 character key, each character must be unique: ')
            good_key = True
            for i in user_key:
                
                #Checking if the input is 36 charcters long
                if len(user_key) != 36:
                    print('Must be 36 characters long!')
                    good_key = False
                    break
                    
                #Checking if the input only contains unique values
                elif user_key.count(i) >1:
                    print('Must contain only unique values!')
                    good_key = False
                    break
            
            #Stores the input in final_key when it meets all critera
            if good_key == True:
                final_key = user_key
                return(final_key)
            
    #Runs if the user inputs something other than 1 or 2 in user_opt
    else:
        print('Invalid Input')
        
#Uses a key from the grab_key function to either encode or decode a user input message
def codinator():
    
    #Asks the user to input a 1 or 2 to either encode or decode a message
    enco_deco = input('Please enter 1 for encoding or 2 for decoding: ')
    
    #Runs the grab_key function and stores the final_key
    final_key = grab_key()
    
    #Asks the user to input a message to be encoded or decoded, in which there should be no punctuation
    user_mes = input('Please enter a message to be encoded (no punctuation) or decoded: ')
    
    #Creates a blank string for storing the end result
    enco_mes = ''
    
    #Runs if the user inputs 1 in enco_deco, sets the 'from' reference to be alph_ref
    #and sets the 'to' refernce to be final_key
    if enco_deco == '1':
        co_from = alph_ref
        co_to = final_key
        
    #Runs if the user inputs 2 in enco_deco, does the inverse of the abouve if statement
    elif enco_deco == '2':
        co_from = final_key
        co_to = alph_ref
        
    #Tells the user their input is invalid if they input something other than 1 or 2 in enco_deco
    else:
        print('Invalid Input')
        
    #Sets a for loop going through the index numbers of each character in the user input message
    for inde in range(len(user_mes)):
        
        #Stores the specific character at the current index in the user message
        char = user_mes[inde]
        
        #Checks if the character is a letter in the alphabet, if so checks if it is uppercase
        #If it is uppercase, sets it to lowercase and adds the special character '‼' to the end message
        #The special character is used later to determine uppercase upon decoding
        if char.isalpha() == True:
            if char.isupper() == True:
                char = char.lower()
                enco_mes += '‼'
                
        #Adds a space to the end message where there is a space in the user message
        if char == ' ':
            enco_mes += ' '
            
        #Adds nothing to the end message where there is the '‼' character
        elif char == '‼':
            enco_mes += ''
            
        #Checks if the charater before the current index is '‼'
        #If so makes the added character uppercase
        elif user_mes[inde-1] == '‼':
            
            #Gets the translation index by finding the same character in the 'from' reference
            #and returning its index
            tran_inde = co_from.rfind(char)
            
            #Uses the translation index to find a character in the 'to' reference,
            #adding it to the end message
            enco_mes += (co_to[tran_inde].upper())
            
        #Does the same as the above elif statement, except doesn't set the added character
        #to be uppercase
        else:
            tran_inde = co_from.rfind(char)
            enco_mes += (co_to[tran_inde])
            
    #Prints the end message for the user
    print('Message: ' + enco_mes)

#Sample Keys:
#Set 1: qwertyuiopasdfghjklzxcvbnm0987654321
#Set 2: /*-.0142thsfnk';zxcvbqmlpoiu9876\=+_
#Set 3: $wb)2:?\,]vs[#&i}or`71cf9"nk*<8ye{!>

#Sample Messages for Decoding:
#Set 1: ‼ziqfa ngx ygk sggaofu qz dn egrt iqct q uktqz rqn
#Set 2: ‼2';0 p'bx ./p tc 4'tk4 m0ff v2/ksc 1'x f''stk4 /v np -'.0
#Set 3: ‼\&i2 9&7 2#]&92) [9 s,``s2 io&]2b` ‼io&ir `& [9 o&&[[$`2 :&o )2$s,#? c,`\ [9 o$[ws,#?

codinator()
```
